```python
import os
import pandas as pd
folder_path = '/content/drive/MyDrive/הדסים/finak'
files_with_new = [f for f in os.listdir(folder_path) if os.path.isfile(os.path.join(folder_path, f)) and 'new' in f]
file_count = len(files_with_new)
print(files_with_new)
print(file_count)
for i in range(1, file_count):
    df = pd.read_csv(f'/content/drive/MyDrive/הדסים/new{i}.csv')
    df['value'] = pd.to_numeric(df['value'], errors='coerce')
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    df['hour'] = df['timestamp'].dt.hour
    hourly_mean = df.groupby('hour')['value'].mean().reset_index()
    hourly_mean['file_id'] = i
    hourly_mean['date'] = df['timestamp'].dt.date.iloc[0] if not df.empty else None
    hourly_mean.to_csv(f'{folder_path}/hourly_mean_{i}.csv', index=False)
final_file = pd.DataFrame()
for i in range(1, file_count):
     hourly_df = pd.read_csv(f'{folder_path}/hourly_mean_{i}.csv')
     final_file = pd.concat([final_file, hourly_df], ignore_index=True)
```

```python
print(final_file)
```

```python
_df = final_file.groupby(['date', 'hour'], as_index=False)['value'].mean()

_df['שעת'] = _df['hour'].apply(lambda x: f'{x:02}:00')

_df = _df[['date', 'שעת', 'value']]
_df.columns = ['ממוצע', 'שעת', 'התחלה']

print(_df)
```
