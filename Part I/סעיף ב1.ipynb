```python
df=pd.read_csv('/content/drive/MyDrive/בינה מלאכותית/time_series.csv')
```

```python
df.dropna(inplace=True)
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
df['date'] = df['timestamp'].dt.date.astype(str)  # שומר את התאריך בפורמט YYYY-MM-DD
df['hour'] = df['timestamp'].dt.strftime('%H:00')  # שומר את השעה בפורמט HH:00
_df = df.groupby(['date', 'hour'], as_index=False)['value'].mean()
print(_df)
```

```python

_df = _df.rename(columns={"datetime": "תאריך ושעה", "value": "ממוצע"})
```
